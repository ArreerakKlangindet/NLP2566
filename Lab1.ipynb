```python
# {'คำถาม':'คำตอบ'}

data = lines[1:]

qa_dict = {}

for item in data:
  x = item.split("\t")
  #print(x[0],x[1])
  words = word_tokenize(x[0].replace("?"," ").strip(), engine="newmm")

  seg_w = " ".join(words)
  qa_dict[seg_w] = x[1]
```

```python

while True:
  _max = 0
  max_q = ''
  max_ans = ''
  user_text = input("ถามไรตอบได้ : ")
  if user_text == "exit":
    break
  _tmp = word_tokenize(user_text)
  user_q = " ".join(_tmp)

  for key in qa_dict.keys():
    #print(key)

    text = key.replace("?","").lower()
    uq = user_q

    words = text.split(" ")
    uq_words = uq.split(" ")
    #print(words)
    #print(uq_words)
    count = 0

    for w in uq_words:
      if w in words:
        count +=1
    if _max < count:
      _max = count
      max_q = key
      max_ans = qa_dict[key]

  print(max_q)
  print(max_ans)
  print(_max)
```
